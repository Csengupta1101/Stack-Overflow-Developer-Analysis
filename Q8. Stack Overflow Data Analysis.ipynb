```python
all_columns=P_Data.select_dtypes(include=['object','int64','float64']).columns.tolist()
for column in P_Data:
    if P_Data[column].isnull().any():
        if(column in all_columns):
            P_Data[column]=P_Data[column].fillna(P_Data[column].mode()[0])
                
```
